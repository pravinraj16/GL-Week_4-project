# 3. Display the percentage distribution of the data in each year using the pie chart.

```python
years = df.groupby('year').size()/len(df)*100
plt.figure(figsize=(6,8))
plt.pie(years, labels=years.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year')
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
plt.figure(figsize = (6,8))
sns.distplot(df['priceperlb'])
plt.title("Distribution of the variable (price per lb) using displot")
plt.show()
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and
write your findings.

```python
plt.figure(figsize=(6,8))
plt.scatter(df['numcol'], df['prodvalue'],color = "green")
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')
plt.title('Relationship between Number of Colonies and Production Value using a Scatter Plot.')
plt.show()
```

Findings:
The distribution of the variable (price per lb) using displot shows that the average price per lb for a given item is between $1 and $10. The distribution is slightly right-skewed, with a few outliers at the higher end. There is also a small peak at around $5 per lb, which may suggest that this is a common price point for the item. 

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable
'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(9,9))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Relationship between Year and Prodvalue')
plt.show()
```

Findings:
The boxplot shows that the median prodvalue has increased over the years. The median prodvalue in 1998 was around 2 and the median prodvalue in 2012 was around 6. This suggests that there is a positive relationship between year and prodvalue.

# Q7) Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences.

```python
sns.pairplot(df, vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year'], hue='year',palette='viridis')
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol',
'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
data_subset = df[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']]
corr_matrix = data_subset.corr()
sns.heatmap(corr_matrix, annot=True,cmap = "BuPu")
plt.title("Correlation Matrix")
plt.show()
```
